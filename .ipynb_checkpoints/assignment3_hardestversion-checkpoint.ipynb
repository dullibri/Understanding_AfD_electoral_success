In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

**Harder option:** Implement the bar coloring as described in the paper, where the color of the bar is actually based on the amount of data covered (e.g. a gradient ranging from dark blue for the distribution being certainly below this y-axis, to white if the value is certainly contained, to dark red if the value is certainly not contained as the distribution is above the axis).

**Even Harder option:** Add interactivity to the above, which allows the user to click on the y axis to set the value of interest. The bar colors should change with respect to what value the user has selected.

**Hardest option:** Allow the user to interactively set a range of y values they are interested in, and recolor based on this (e.g. a y-axis band, see the paper for more details).

```python
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.animation as animation
from scipy import stats
from matplotlib import cm
from matplotlib.widgets import SpanSelector
```

```python
# Prepare data 
np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])


describe = pd.DataFrame({'Mean':df.T.mean(), 'Std': df.T.std()})
describe['Error'] = describe.Std/np.sqrt(len(df.T))*1.96
describe['Upper'] = describe.Mean+describe.Error
describe['Lower'] = describe.Mean-describe.Error


#maler(df.iloc[0,:], 80000,120000)
```

```python
def maler(describe_i, dist, ymin, ymax, cmp):
    '''
    Provides the color for each bar. 
    
    There are three constellations: If the mean is included in the range provided
    by the users, it returns white. If not, it can either be above or below. Than,
    the closest y-value, i.e. either ymin (if above) or ymax (if below) is tested to 
    be the mean of the respective distribution (=Null hypothesis).
    
    The p-value is mapped into colormap 'seismic'. This map ranges from dark blue (value: 0)
    to white (value:0.5) and finally dark red (value: 1). Thus if the mean of the distribution
    is above the range it needs to become red. The more strongly the hypothesis of ymax being
    the mean of the distribution is rejected, the lower the p-value. However, seismic's values
    turn more red, the HIGHER the value in the range [0,1]. Thus, -2 is deducted and divided by
    -1. And, as seismic is red only from [0.5, 1] it is divided by 2. 
    
    If the ymin is above, remapping is easier as it only has to be mapped into the range
    [0.5, 1], thus, it is divided by 2.
    
    Paramaters
    -----------
    describee_i: pd.DataSeries
        contains mean, std, error, upper and lower bound of 95%-confidence intervall
    ymin, ymax: int or float
        The lower and the upper boundery of the range provided by the user
    cmp: colormap
    
    Returns
    -------
    color:
        color value for use in barchart            
      
    '''
    mean, std, error, upper, lower = describe_i
    
    # range includes the mean
    if (ymin<=mean)&(mean<=ymax):
        colval = 0.5
    # range is above mean
    elif (ymin>mean):
        t, p = stats.ttest_1samp(dist, ymin)
        colval = p/2 
    elif (ymax<mean):
        t, p = stats.ttest_1samp(dist, ymax)
        # rescale into 0 to 0.5
        colval = (p-2)/-2              
    color = cmp(colval)
    return color
```

```python
cmp = cm.get_cmap('seismic')
    
fig = plt.figure()
plt.subplots_adjust(right=0.85, top=0.85)

ax1 = plt.subplot(111)
ax1.set_title('Are the means within a given range?\nCheck with your mouse.',pad=20)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)

ind = range(1,5)
ymax_start=42000
ymin_start=40000


color_new = [maler(describe.iloc[i]
                   ,dist=df.iloc[i,:]
                   ,ymin=ymin_start
                   ,ymax=ymax_start
                   ,cmp=cmp) for i in range(len(describe))]


bar = ax1.bar(ind
              , describe.Mean
              , color=color_new
              , edgecolor='black'
              , yerr=describe.Error
              , ecolor='black'
              , align='center'
              , alpha=1
              , capsize=10
              , tick_label=df.index
             )
# as errorbars disappear when updating, hence the lines are ploted separatly
for i in range(4):
    ax1.plot([i+1, i+1],[describe.Lower.iloc[i]
                       ,describe.Upper.iloc[i]]
             ,color='black')

# set and format labels and ticks
ax1.set_xticks([1,2,3,4])
ax1.set_xlabel('Sample')
ax1.set_ylabel('Mean')
y_labels = ax1.get_yticks()
ax1.set_yticklabels(['{:,.0f}'.format(y) for y in y_labels])

# plot initial y-range
lmax,  = ax1.plot([0.2,4.5],[ymax_start]*2, color='red',alpha=0.4)
lmin,  = ax1.plot([0.2,4.5],[ymin_start]*2, color='red',alpha=0.4)
shade = ax1.fill_between([0.2,4.5],
                 [ymax_start]*2,
                 [ymin_start]*2,
                 facecolor='red',
                 alpha=0.4
                )

# Colorbar
sm = cm.ScalarMappable(cmap=cmp)#, norm=plt.Normalize(0,.5))
sm.set_array([])

cbar = plt.colorbar(sm)
cbar.set_ticks([0.025,0.5,(1-0.025)])
cbar.set_ticklabels(['below\n(5% level)'
                     ,'in range'
                     ,'mean is above\n(5% level)'])

# define interactive function to be called by Spanselector
def onselect(ymin, ymax):
    ax1.collections.clear()
   
    color_update = [maler(describe.iloc[i]
                          ,dist=df.iloc[i,:]
                          ,ymin=ymin
                          ,ymax=ymax
                          ,cmp=cmp) for i in range(len(describe))]

    for i in range(4):
        ax1.patches[i].set_color(color_update[i])
        ax1.patches[i].set_edgecolor('black')

    lmin.set_ydata([ymin]*2)

    lmax.set_ydata([ymax]*2)
    shade = ax1.fill_between([0.2,4.5],
                 [ymax]*2,
                 [ymin]*2,
                 facecolor='red',
                 alpha=0.4
                )
           
span = SpanSelector(ax1, onselect, 'vertical', useblit=True,
                    rectprops=dict(alpha=0.5, facecolor='red'))

```
